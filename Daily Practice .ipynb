```python
# Write a Python program that takes a sentence as input and prints the sentence with the words reversed — not the characters.
```

```python
Sentence = input("Enter your sentece here ; ")
Rewarse_Sentence = ' '.join(Sentence.split()[::-1])
print(Rewarse_Sentence)
```

```python
# Write a Python program that checks whether a given string is a palindrome (reads the same forward and backward).
```

```python
Word = input("Enter your word in lowercase: ")
Word = Word.lower()
if Word == Word[::-1]:
    print(f'{Word} is palindrome')
else:
    print(f'{Word} is not palindrome')
```
